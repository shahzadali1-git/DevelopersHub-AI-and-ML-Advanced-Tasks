
# News Topic Classifier Using BERT

## Overview
This project is part of the **AI/ML Engineering Advanced Internship at DevelopersHub Corporation**.  
It demonstrates how to **fine-tune a transformer model (DistilBERT)** to classify news headlines into topic categories using the **AG News dataset**.

The notebook includes:  
- Tokenization and preprocessing of the dataset  
- Fine-tuning a transformer model  
- Evaluation using **accuracy** and **F1-score**  
- Live deployment using **Gradio** for interactive predictions  

---

## Dataset
- **AG News Dataset** (Available via Hugging Face Datasets)  
- **Training set:** 120,000 samples  
- **Test set:** 7,600 samples  
- **Classes:**  
  1. World  
  2. Sports  
  3. Business  
  4. Science/Technology
